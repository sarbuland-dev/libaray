```python
import numpy as np

marks = np.random.randint(40, 100, size=(10, 5))  # 10 students × 5 subjects
total = np.sum(marks, axis=1)
avg = np.mean(marks, axis=1)
grades = np.where(avg >= 80, 'A',
          np.where(avg >= 60, 'B',
          np.where(avg >= 40, 'C', 'Fail')))

print("Marks:\n", marks)
print("total marks",total)
print("Average:", avg)
print("Grades:", grades)
```

```python
import numpy as np
import matplotlib.pyplot as plt
marks=np.random.randint(0,100,size=(10,5))
total=np.sum(marks,axis=1)
avg=np.average(marks,axis=1)
grade=np.where(avg>=80,'A',
       np.where(avg>=70,'B',
       np.where(avg>=50,'C','F')))

result=np.where(grade=='F','Fail','Pass')
print("marks are",marks)
print("total are ",total)
print("avg are",avg)
print("grades are",grade)
print("result are ",result)

students = np.arange(1, 11)  # Student numbers 1–10
plt.bar(students, avg, color='green', edgecolor='black')
plt.xlabel('Student Number')
plt.ylabel('Average Marks')
plt.title('Average Marks of Students')
plt.xticks(students)
plt.axhline(50, color='red', linestyle='-', label='Passing Line')
plt.legend()
plt.show()

```

```python
import numpy as np

marks = np.random.randint(40, 100, size=(5, 5))
avg = np.mean(marks, axis=1)
grades = np.where(avg >= 90, 'A+',
          np.where(avg >= 80, 'A',
          np.where(avg >= 70, 'B',
          np.where(avg >= 60, 'C', 'Fail'))))

topper_index = np.argmax(avg)

print("Marks:\n", marks)
print("Average:", avg)
print("Grades:", grades)
print(f"Topper is student #{topper_index+1} with average {avg[topper_index]}")
```
